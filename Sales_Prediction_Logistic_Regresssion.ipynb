## Loading Dataset

I don't have dataset so i am creating some random data.

```python
import random
import pandas as pd

import random
import pandas as pd

# Set the number of rows to generate
num_rows = 300

# Create empty lists to store the data
age = []
salary = []
status = []

# Generate random data for each column
for i in range(num_rows):
    # Determine age group based on age
    if i < num_rows * 0.1:
        age_group = 'child'
        salary_range = (0, 3000)
    elif i < num_rows * 0.3:
        age_group = 'teen'
        salary_range = (3000, 10000)
    elif i < num_rows * 0.6:
        age_group = 'less young'
        salary_range = (10000, 20000)
    elif i < num_rows * 0.8:
        age_group = 'moderate adult'
        salary_range = (20000, 40000)
    else:
        age_group = 'good with increase in age'
        salary_range = (40000, 80000)
    
    # Calculate salary based on age group
    age.append(random.randint(10, 100))
    salary.append(random.randint(*salary_range))
    
    # Determine status based on age group
    if age_group in ['child', 'teen']:
        status.append(0)
    else:
        status.append(1)

# Create a dictionary to store the data
data = {'age': age, 'salary': salary, 'status': status}

# Convert the dictionary to a Pandas DataFrame
df = pd.DataFrame(data)

# Display the first 5 rows of the DataFrame
print(df.head())
```

## Segregating Dataset into X & Y

```python
X = df.iloc[:,:-1].values
X
```

```python
Y = df.iloc[:,-1].values
Y
```

## Splitting Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

## Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

## Predicting For Test Data

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix")
print(cm)
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
```
